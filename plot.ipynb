```python
import matplotlib.pyplot as plt
```

```python
with open("data.txt", "r") as file:
    time = []
    values = []
    for line in file:
        columns = line.strip().split()
        read_time = columns[0].split(':')
        time.append(float(read_time[0])*60+float(read_time[1]))
        values.append(float(columns[1]))

# Plot data
plt.plot(time, values, '-')
plt.xlabel('Time')
plt.ylabel('%MEM')
plt.title('MEM% for connectedcomponent process')
plt.show()
```

```python
with open("pagerank-graphchi.java.txt", "r") as file:
    time = []
    reads = []
    writes = []
    i=0
    for line in file:
        columns = line.strip().split()
        time.append(i)
        reads.append(float(columns[4]))
        writes.append(float(columns[5]))
        i= i+1

# Plot data
plt.plot(time, reads,color='red', label= 'reads')
plt.plot(time, writes,color='blue',label = 'writes')
plt.xlabel('Time')
plt.ylabel('IO')
plt.title('page Read/Write over time')
plt.show()
```
